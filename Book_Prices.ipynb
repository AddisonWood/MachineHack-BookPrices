```python
training_data = pd.DataFrame(pd.read_excel("Participants_Data/Data_Train.xlsx"))

iqr = stats.iqr(training_data["Price"])
outlier_cutoff_high = np.quantile(np.array(training_data["Price"]), .75) + iqr*1.5
print(len(training_data[training_data["Price"] > outlier_cutoff_high]), "outliers")
training_data = training_data[training_data["Price"] < outlier_cutoff_high]
training_data = training_data.reset_index().drop(columns=["index"])

training_data["Reviews_Cleaned"] = 0.1
for i in range(len(training_data["Reviews"])):
    training_data["Reviews_Cleaned"][i] = float(training_data["Reviews"][i][:3])
    
training_data["Year"] = 0.1
for i in range(len(training_data["Edition"])):
    try:
        year = int(training_data["Edition"][i][-4:])
        training_data["Year"][i] = 2020-year
    except ValueError:
        training_data["Year"][i] = 9999
mean_year = int(training_data[training_data["Year"] < 9999]["Year"].mean())
for i in range(len(training_data["Year"])):
    if (training_data["Year"][i] == 9999):
        training_data["Year"][i] = 2020-mean_year
    
training_data["Paperback?"] = 0
for i in range(len(training_data["Edition"])):
    if ("Hardcover" in training_data["Edition"][i]):
        training_data["Paperback?"][i] = 0
    elif ("Paperback" in training_data["Edition"][i]):
        training_data["Paperback?"][i] = 1
    else:
        training_data["Paperback?"][i] = 2

training_data["Ratings_Cleaned"] = 1
for i in range(len(training_data["Ratings"])): 
    num_ratings = int(training_data["Ratings"][i][:training_data["Ratings"][i].find(" ")].replace(",", ""))
    training_data["Ratings_Cleaned"][i] = num_ratings

training_data["Import?"] = 0
for i in range(len(training_data["Import?"])):
    if ("Import" in training_data["Edition"][i]):
        training_data["Import?"][i] = 1
    else:
        training_data["Import?"][i] = 0
        
category_prices = {}
for cat in training_data["BookCategory"].unique():
    med = training_data[training_data["BookCategory"] == cat]["Price"].median()
    category_prices[cat] = med
training_data["Category_Vals"] = 0.1
for i in range(len(training_data["BookCategory"])):
    training_data["Category_Vals"][i] = category_prices[training_data["BookCategory"][i]]
    
training_data["Book?"] = 0
for i in range(len(training_data["Genre"])):
    if ("(Books)" in training_data["Genre"][i]):
        training_data["Book?"][i] = 1
    elif ("Textbooks" in training_data["Genre"][i]):
        training_data["Book?"][i] = 2
    else:
        training_data["Book?"][i] = 0

    
#print(mean_year)
#training_data[training_data["Price"] > outlier_cutoff_high].head(600)
```

```python
testing_data = pd.DataFrame(pd.read_excel("Participants_Data/Data_Test.xlsx"))

testing_data["Reviews_Cleaned"] = 0.1
for i in range(len(testing_data["Reviews"])):
    testing_data["Reviews_Cleaned"][i] = float(testing_data["Reviews"][i][:3])
    
testing_data["Year"] = 1
for i in range(len(testing_data["Edition"])):
    try:
        year = int(testing_data["Edition"][i][-4:])
        testing_data["Year"][i] = 2020-year
    except ValueError:
        testing_data["Year"][i] = 9999
mean_year = int(testing_data[testing_data["Year"] < 9999]["Year"].mean())
for i in range(len(testing_data["Year"])):
    if (testing_data["Year"][i] == 9999):
        testing_data["Year"][i] = 2020-mean_year

testing_data["Paperback?"] = 0
for i in range(len(testing_data["Edition"])):
    if ("Hardcover" in testing_data["Edition"][i]):
        testing_data["Paperback?"][i] = 2
    elif ("Paperback" in testing_data["Edition"][i]):
        testing_data["Paperback?"][i] = 1
    else:
        testing_data["Paperback?"][i] = 0

testing_data["Ratings_Cleaned"] = 1
for i in range(len(testing_data["Ratings"])): 
    num_ratings = int(testing_data["Ratings"][i][:testing_data["Ratings"][i].find(" ")].replace(",", ""))
    testing_data["Ratings_Cleaned"][i] = num_ratings

testing_data["Import?"] = 0
for i in range(len(testing_data["Import?"])):
    if ("Import" in testing_data["Edition"][i]):
        testing_data["Import?"][i] = 1
    else:
        testing_data["Import?"][i] = 0

testing_data["Category_Vals"] = 0.1
for i in range(len(testing_data["BookCategory"])):
    testing_data["Category_Vals"][i] = category_prices[testing_data["BookCategory"][i]]
    
testing_data["Book?"] = 0
for i in range(len(testing_data["Genre"])):
    if ("(Books)" in testing_data["Genre"][i]):
        testing_data["Book?"][i] = 1
    elif ("Textbooks" in testing_data["Genre"][i]):
        testing_data["Book?"][i] = 2
    else:
        testing_data["Book?"][i] = 0

        
```

```python
#X_train = np.array(training_data[["Reviews_Cleaned", "Year", "Paperback?"]])
#y_train = np.array(training_data["Price"])

#X_test = np.array(testing_data[["Reviews_Cleaned", "Year", "Paperback?"]])
#y_test = np.array(testing_data["Price"])

X = np.array(training_data[["Reviews_Cleaned", "Year", "Paperback?", "Ratings_Cleaned", "Import?", "Category_Vals", "Book?"]])
y = np.array(training_data["Price"])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
params = {'n_estimators': 500,
          'max_depth': 3,
          'min_samples_split': 5,
          'learning_rate': 0.05,
          'loss': 'ls'}


linear_regression = LinearRegression(normalize=True)
linear_regression.fit(X_train,np.log(y_train))
y_pred = np.exp(linear_regression.predict(X_test))
regression_score = 1 - np.sqrt(np.square(np.log10(y_pred+1)-np.log10(y_test+1)).mean())
print("Normal Regression: ", regression_score)

model = ensemble.GradientBoostingRegressor(**params)
model.fit(X_train, y_train)
y_pred2 = model.predict(X_test)
xgboost_score = 1 - np.sqrt(np.square(np.log10(y_pred2+1)-np.log10(y_test+1)).mean())
print("XGBoost Regression:", xgboost_score)
```
